```python
print("Bilangan Genap:")
for i in range(1, 51):
  if i % 2 == 0:
    print(i)

print("\nBilangan Ganjil:")
for i in range(1, 51):
  if i % 2 != 0:
    print(i)
```

```python
print("Bilangan Prima antara 1 sampai 100:")
for angka in range(2, 101):
  is_prima = True
  for i in range(2, angka):
    if (angka % i) == 0:
      is_prima = False
      break
  if is_prima:
    print(angka)
```
