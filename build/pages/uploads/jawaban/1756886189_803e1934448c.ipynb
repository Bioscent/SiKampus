```python
#input value ke dalam variable
panjang = 12
lebar = 5
tinggi = 8

#hitung luas, volume, dan keliling
luas = panjang * lebar
volume = panjang * lebar * tinggi
keliling = 2 * (panjang + lebar)

print("luas =", luas)
print("volume =", volume)
print("keliling =", keliling)

#cek menggunakan operator sintaks
print("Apakah luas bangunan tersebut lebih luas dari 50?", luas > 50)
print("Apakah volume tersebut bernilai 480?", volume == 480)
```
